```python
import os
import shutil
from PIL import Image
import pandas as pd
import seaborn as sns
```

## Young Klasörü

```python
src = r"C:\Users\CASPER\Desktop\Dönem 2\Bitirme Projesi\Dataset\UTKFace Aging\UTKFace"
dest = r"C:\Users\CASPER\Desktop\Dönem 2\Bitirme Projesi\Dataset\UTKFace Aging\Young"
src_files = os.listdir(src)
i = 1
young = list()
for file_name in src_files:
    full_file_name = os.path.join(src, file_name)
    im = Image.open(full_file_name)
    if im.mode == 'RGB':
        age = file_name.split('_')
        age = int(age[0])
        
        if 20< age <30:
            i+=1
            young.append(age)
            if os.path.isfile(full_file_name):
                shutil.copy(full_file_name, dest)
        if i > 2100:
            break
```
